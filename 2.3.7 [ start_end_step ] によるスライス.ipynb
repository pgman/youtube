# **2.3.7 [ start:end:step ] によるスライス**

*   [:]は、先頭から末尾までのシーケンス全体を抽出する。
*   [start:]は、startオフセットから末尾までのシーケンスを抽出する。
*   [:end]は、先頭からend-1オフセットまでのシーケンスを抽出する。
*   [start:end]は、startオフセットからend-1オフセットまでのシーケンスを抽出する。
*   [start:end:step]は、step文字ごとにstartオフセットからend-1オフセットまでのシーケンスを抽出する。

```
letters = 'abcdefghijklmnopqrstuvwxyz'
```

:だけ指定すると文字全体が表示される

```
letters[:]
```

オフセット20から末尾まで切り取る例


```
letters[20:]
```

オフセット10から末尾まで切り取る例

```
letters[10:]
```

オフセット12から14まで切り取る例<br>
(Pythonは最後のオフセットを部分文字列に入れないことに注意しよう)

```
letters[12:15]
```

最後の3文字を取り出す

```
letters[-3:]
```

次のサンプルでは、オフセット18から末尾より4文字手前までを切り取る。前の例では-3でxを取り出していたが、今度は-3が実際には-4まで進んでwで止まることに注意しよう。

```
letters[18:-3]
```

末尾6文字手前から末尾の3文字手前まで

```
letters[-6:-2]
```

先頭から末尾まで7字ごとに取り出す

```
letters[::7]
```

オフセット4から19まで、3文字ごとに取り出す

```
letters[4:20:3]
```

オフセット19から末尾まで、4文字ごとに取り出す

```
letters[19::4]
```

先頭からオフセット20まで5文字ごとに取り出す

```
letters[:21:5]
```

ステップサイズに負数を指定すると、逆にステップしていく。

```
letters[-1::-1]
```

次のようにしても同じ結果が得られる

```
letters[::-1]
```

スライスは間違ったオフセットに寛容だ

```
letters[-50:]
```

末尾の51字手前から末尾50字手前まで

```
letters[-51:-50]
```

先頭から70字までを指定するとこうなる

```
letters[:70]
```

70字目から70字目までを指定すると空文字列になる

```
letters[70:71]
```
